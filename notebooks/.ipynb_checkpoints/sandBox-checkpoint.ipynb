```python
def make_dir(fix_path, ano,mes,dia):
    if os.path.isdir(f'{fix_path}/{ano}'):
        pass
    else:
        os.mkdir(f'{fix_path}/{ano}')

    if os.path.isdir(f'{fix_path}/{ano}/{mes}'):
        pass
    else:
        os.mkdir(f'{fix_path}/{ano}/{mes}')

    if os.path.isdir(f'{fix_path}/{ano}/{mes}/{dia}'):
        pass
    else:
        os.mkdir(f'{fix_path}/{ano}/{mes}/{dia}')
        
def get_url(poder,ano,mes,dia,pg):
    
    if pg < 10:
        url = f'http://diariooficial.imprensaoficial.com.br/doflash/prototipo/{ano}/{mes}/{dia}/{poder}/pdf/pg_000{pg}.pdf'
    elif (pg>=10) & (pg<100):
        url = f'http://diariooficial.imprensaoficial.com.br/doflash/prototipo/{ano}/{mes}/{dia}/{poder}/pdf/pg_00{pg}.pdf'
    elif (pg>=100) & (pg<1000):
        url = f'http://diariooficial.imprensaoficial.com.br/doflash/prototipo/{ano}/{mes}/{dia}/{poder}/pdf/pg_0{pg}.pdf'
    elif pg>=1000:
        url = f'http://diariooficial.imprensaoficial.com.br/doflash/prototipo/{ano}/{mes}/{dia}/{poder}/pdf/pg_{pg}.pdf'
        
    return url

def get_month_name(month):
    mes_dict = {
        '01':'Janeiro',
        '02':'Fevereiro',
        '03':'Março',
        '04':'Abril',
        '05':'Mario',
        '06':'Junho',
        '07':'Julho',
        '08':'Agosto',
        '09':'Setembro',
        '10':'Outubro',
        '11':'Novembro',
        '12':'Dezembro',
    }
    
    return mes_dict[month]

def merge_pdf(fix_path,ano, mes_number, dia):          
    path = f'{fix_path}/{ano}/{mes_number}/{dia}'
    pages_dict = {}
    pages = os.listdir(path)
    pages = [page for page in pages if page.split('.')[1] == 'pdf']
    for page in pages:
         pages_dict[int(page.split('_')[1].split('.')[0])] = page

    page_values = list(pages_dict.keys())
    page_values.sort()

    merger = PdfFileMerger()
    for page in page_values:

        merger.append(f'{path}/{pages_dict[page]}')

    merger.write(f"{fix_path}/{ano}/{mes_number}/{ano}-{mes_number}-{dia}.pdf")
    merger.close()

    shutil.rmtree(path, ignore_errors=True)
```

```python
import requests
import pandas as pd
import datetime
import os
import shutil
from PyPDF2 import PdfFileMerger
```

```python
d=1

#opcoes legislativo , exec1, exec2
poder = "exec1"

fix_path =f'../data/pdf/{poder}'

datelist = pd.date_range(datetime.datetime.today() - datetime.timedelta(days=d), datetime.datetime.today()).tolist()

df = pd.read_csv(f'{fix_path}/controle_{poder}.csv')
controle_dates = df['date'].to_list()

download_dates = [date for date in datelist if date.strftime('%d/%m/%Y') not in controle_dates]
```

```python

for date in download_dates:
    
    ano = date.strftime('%Y')
    mes_name = get_month_name(date.strftime('%m'))
    mes_number = date.strftime('%m')
    dia = date.strftime('%d')
    
    pg=1
    status_code = 200
    make_dir(fix_path,ano,mes_number,dia)

    while status_code == 200:

        url =  get_url(poder,ano,mes,dia,pg)
#         print(url)
        
        r = requests.get(url, stream=True)
        status_code = r.status_code
        
        if status_code == 200:
            with open(f'{fix_path}/{ano}/{mes_number}/{dia}/pg_000{pg}.pdf', 'wb') as file:
                file.write(r.content)
            pg+=1

        else:
            arquivo = 0
            pg-=1
            pass

        
    url =  get_url(poder,ano,mes,dia,pg)
    
    if (arquivo==0) & (pg==0):
        existe = 'nao'
        exist  = 'nao existe'
        
        shutil.rmtree(f'{fix_path}/{ano}/{mes_number}/{dia}', ignore_errors=True)
        
    else:
        existe = 'sim'
        exist  = 'existe'
        
        merge_pdf(fix_path,ano, mes_number, dia)
        
        
        
        
        
        
        
    
    data = date.strftime('%d/%m/%Y')
    print(f'Diario do dia {data}, {exist} e tem {pg} paginas' '  -  ', url)
    
    dd = pd.DataFrame([date.strftime('%d/%m/%Y')], columns=['date'])
    dd['diario_existe'] = existe
    dd['paginas'] = pg
    
    dd.to_csv(f'{fix_path}/controle_{poder}.csv', index=False, mode='a', header=False)
```

# Searching

```python
import pdftotext

def find_in_pdf(path, string):
    with open(path, "rb") as f:
        pdf = pdftotext.PDF(f)
        
    pages = {}
    
    for i in range(len(pdf)):
    #     print(i,'\n', pdf[i])
        content = pdf[i].replace('(','').replace(')','').replace('\n','')
        search = content.count(string)

        if search!=0:
            pages[i+1] = search
#         else:
#             pages[i] = 0
    return pages
```

```python
poder = 'exec1'
ano   = 2020
mes_number = '04' 

path = f'../data/pdf/{poder}/{ano}/{mes_number}'
files = os.listdir(path)
pdf_files = [file for file in files if file.split('.')[1] == 'pdf']


```

```python

strings = ['Kamilos']

df_final = pd.DataFrame()

for pdf_file in pdf_files:
    for string in strings:
        pdf_file_path = f'{path}/{pdf_file}'
        pages = find_in_pdf(pdf_file_path,string)

        df = pd.DataFrame(pages.keys(), columns=['pagina'])
        df['contagem'] = pages.values()
        df['busca']    = string
        df['arquivo']  = pdf_file
        cols = ['arquivo', 'busca', 'contagem', 'pagina']
        
        
        df_final = pd.concat([df_final,df[cols]],0)
```
